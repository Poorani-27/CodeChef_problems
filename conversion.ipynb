```python
def decimal_t0_binary(n):
    res = bin(n)[2:]
    return res

print(decimal_t0_binary(18))
print(decimal_t0_binary(15))
```

```python
def binary_to_octal(n):
    decimal=int(n,2)
    octal=oct(decimal)[2:]
    return octal
print(binary_to_octal("1100110"))
print(binary_to_octal("11111"))
```

```python
def octal_to_binary(n):
    decimal=int(n, 8)
    binary=format(decimal,'017b')
    return binary

print(octal_to_binary('170'))
print(octal_to_binary('345'))

    
```

```python
def digits_to_word(n):
    if n==0: return "Zero"
    
    
    units = ["", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
    teens = ["", "eleven", "twelve", "thirteen", "fourteen", "fifteen", "sixteen", "seventeen", "eighteen", "nineteen"]
    tens = ["", "ten", "twenty", "thirty", "forty", "fifty", "sixty", "seventy", "eighty", "ninety"]
    thousands = ["", "thousand", "million", "billion"]
    
    words=""
    check_index=0
    while n > 0:
        if n %1000 !=0:
            chunk_words = ""
            hundreds = n % 1000 // 100
            tens_units = n % 100

            if hundreds != 0:
                chunk_words += units[hundreds] + " hundred"

            if tens_units != 0:
                if hundreds != 0:
                    chunk_words += " "

                if 0 < tens_units < 10:
                    chunk_words += units[tens_units]
                elif 10 <= tens_units < 20:
                    chunk_words += teens[tens_units % 10]
                else:
                    chunk_words += tens[tens_units // 10]
                    if tens_units % 10 != 0:
                        chunk_words += " " + units[tens_units % 10]

            if check_index > 0:
                chunk_words += " " + thousands[check_index]

            words = chunk_words + " " + words

        n //= 1000
        check_index += 1

    return words.strip()
print(digits_to_word(1356))
```
