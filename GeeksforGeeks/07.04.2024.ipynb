```python
class Solution:
    def reciprocalString(self, S):
        result = []

        for char in S:
            if 'A' <= char <= 'Z':
                # Calculate the reciprocal for uppercase letters
                reciprocal_char = chr(ord('Z') - (ord(char) - ord('A')))
            elif 'a' <= char <= 'z':
                # Calculate the reciprocal for lowercase letters
                reciprocal_char = chr(ord('z') - (ord(char) - ord('a')))
            else:
                # Non-letter characters remain the same
                reciprocal_char = char
            
            result.append(reciprocal_char)

        return "".join(result)

# Example usage
sol = Solution()
print(sol.reciprocalString("ab C"))  # Output: "zy X"
print(sol.reciprocalString("Ish"))   # Output: "Rhs"
```

```python
def scores(A, B):
    # Initialize scores for A and B
    ca = 0
    cb = 0
    
    # Iterate over the skills
    for i in range(3):
        if A[i] > B[i]:
            ca += 1
        elif A[i] < B[i]:
            cb += 1
    
    # Return the final scores
    return ca, cb

# Test cases
A1 = [4, 2, 7]
B1 = [5, 6, 3]
result1 = scores(A1, B1)
print(result1)  # Output should be (1, 2)

A2 = [4, 2, 7]
B2 = [5, 2, 8]
result2 = scores(A2, B2)
print(result2)  # Output should be (0, 2)
```
