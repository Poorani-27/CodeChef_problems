```python
s="geeks for geeks"
vowels=['a','e','i','o','u']
for element in s:
    if element in vowels:
        s=s.replace(element,"")
print(s)
```

```python
def splitString(S): 
    alpha=[]
    numeric=[]
    symbols=[]
    for elements in S :
        if elements.isalpha():
            alpha.append(elements)
        elif elements.isnumeric():
            numeric.append(elements)
        else:
            symbols.append(elements)
    print("".join(alpha),end="\n")
    print("".join(numeric),end="\n")
    print("".join(symbols))
    
splitString("12jrleke8***2R2@@4")
```
