```python
def push_choco(N,arr):
    res=[]
    for a in arr:
        if a == 0 :
            res.append(a)
            arr.remove(a)
    return arr+res

N=8
arr=[4,5,0,1,9,0,5,0]
sol=push_choco(N,arr)
print(sol)
```

```python
def push_choco(N,arr):
    count=arr.count(0)
    res=[x for x in arr if x != 0]
    res.extend([0]*count)
    return res
N=8
arr=[4,5,0,1,9,0,5,0]
sol=push_choco(N,arr)
print(sol)
```

```python

def conversion(n):
    binary =bin(n)[2:]
    toggle= "".join('1' if bit == '0' else '0' for bit in binary)
    integer=int(toggle,2)
    return integer

print(conversion(100))
    
    
    
```
